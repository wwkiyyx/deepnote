```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
color = cv2.imread("/work/wangwenkai40/OpenCV/image/lena.png", cv2.IMREAD_COLOR)
showColor(color)
```

```
img = cv2.resize(color, (20, 20))
showColor(img)
```

```
img = cv2.resize(color, (0, 0), fx=0.2, fy=0.4, interpolation=cv2.INTER_AREA)
showColor(img) #fx, fy	沿x轴，y轴的缩放系数
```
