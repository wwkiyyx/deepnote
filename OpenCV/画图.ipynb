```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
color = cv2.imread("/work/f103/coin_7_6x5.jpg", cv2.IMREAD_COLOR)
cv2.line(color, (0, 0), (300, 300), (0, 255, 0), 5) # 图像 起点 终点 颜色 粗细 线型
showColor(color)
```

```
color = cv2.imread("image/lena.png", cv2.IMREAD_COLOR)
cv2.rectangle(color, (10, 10), (300, 300), (0, 255, 0), 5) # 图像 起点 终点 颜色 粗细 线型
showColor(color)
```

```
color = cv2.imread("image/lena.png", cv2.IMREAD_COLOR)
cv2.rectangle(color, (10, 10), (300, 300), (0, 255, 0), -1) # 图像 起点 终点 颜色 粗细 线型
showColor(color)
```

```
color = cv2.imread("image/lena.png", cv2.IMREAD_COLOR)
cv2.circle(color, (300, 300), 100, (0, 255, 0), 5) # 图像 圆心 半径 颜色 粗细 线型
showColor(color)
```

```
color = cv2.imread("image/lena.png", cv2.IMREAD_COLOR)
cv2.circle(color, (300, 300), 100, (0, 255, 0), -1) # 图像 圆心 半径 颜色 粗细 线型
showColor(color)
```
