```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
gray = cv2.imread("image/lena.png", cv2.IMREAD_GRAYSCALE)
canny_edge = cv2.Canny(gray, threshold1=100, threshold2=200)
showGray(canny_edge)
```

```
import numpy as np
gray = cv2.imread("image/lena.png", cv2.IMREAD_GRAYSCALE)
sobel_edge = cv2.Sobel(gray,ddepth=cv2.CV_32F,dx=1,dy=1,ksize=5)
sobel_edge = np.abs(sobel_edge)
sobel_edge = sobel_edge/np.max(sobel_edge)
sobel_edge = sobel_edge*255
sobel_edge = sobel_edge.astype(np.uint8)
showGray(sobel_edge)
```

```
import numpy as np
gray = cv2.imread("image/lena.png", cv2.IMREAD_GRAYSCALE)
sobel_edge_x = cv2.Sobel(gray,ddepth=cv2.CV_32F,dx=1,dy=0,ksize=5)
sobel_edge_x = np.abs(sobel_edge_x)
sobel_edge_x = sobel_edge_x/np.max(sobel_edge_x)
sobel_edge_x = sobel_edge_x*255  #
sobel_edge_x = sobel_edge_x.astype(np.uint8)
sobel_edge_y = cv2.Sobel(gray,ddepth=cv2.CV_32F,dx=0,dy=1,ksize=5)
sobel_edge_y = np.abs(sobel_edge_y)
sobel_edge_y = sobel_edge_y/np.max(sobel_edge_y)
sobel_edge_y = sobel_edge_y*255
sobel_edge_y = sobel_edge_y.astype(np.uint8)
sobel_edge1 = cv2.addWeighted(sobel_edge_x,0.5,sobel_edge_y,0.5,0)
showGray(sobel_edge1)
```

```
import numpy as np
gray = cv2.imread("image/lena.png", cv2.IMREAD_GRAYSCALE)
dst_img_gray = cv2.Laplacian(gray, cv2.CV_32F)
laplacian_edge_gray = cv2.convertScaleAbs(dst_img_gray)
showGray(laplacian_edge_gray)
```
