```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
import numpy as np
mark = np.zeros((300, 300), np.uint8)
x = 150
y = 150
cv2.rectangle(mark, (x-10,y-50), (x+10,y+50), 255, -1)
cv2.rectangle(mark, (x-50,y-10), (x+50,y+10), 255, -1)
showGray(mark)
```

```
roi = mark[100:200,100:200]
showGray(roi)
```

```
res = cv2.matchTemplate(mark, roi, cv2.TM_SQDIFF)
dst = cv2.normalize(res,0,255,cv2.NORM_MINMAX)
showGray(dst)
```

```
res = cv2.matchTemplate(mark, roi, cv2.TM_SQDIFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
max_val
```

```
color = cv2.imread("/work/bp/FOG_4_source.jpg", cv2.IMREAD_GRAYSCALE)
color.shape
```

```
showGray(color)
```

```
roi = color[30:170,500:1000]
showGray(roi)
```

```
res = cv2.matchTemplate(color, roi, cv2.TM_SQDIFF)
dst = cv2.normalize(res,0,255,cv2.NORM_MINMAX)
showGray(dst)
```

```
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
min_val
```

```
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = min_loc
w, h = roi.shape[::-1]
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(color,top_left, bottom_right, 255, 5)
showGray(color)
```
