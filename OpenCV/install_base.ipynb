```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.figure(figsize=(10,10))
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.figure(figsize=(10,10))
    plt.imshow(mat_show)
```

```
import numpy as np
import os
black1 = np.zeros((300, 300), np.uint8)
black3 = np.zeros((300, 300, 3), np.uint8)
gray1 = cv2.add(black1, 128)
gray3 = cv2.add(black3, 128)
white1 = black1 + 255
white3 = black3 + 255
cv2.imwrite("tmp1.bmp", white1)
cv2.imwrite("tmp3.bmp", white3)
mat1 = cv2.imread("tmp1.bmp", cv2.IMREAD_GRAYSCALE)
mat3 = cv2.imread("tmp3.bmp", cv2.IMREAD_COLOR)
os.remove("tmp1.bmp")
os.remove("tmp3.bmp")
print('shape : ', mat3.shape)
print('size : ', mat3.size )
print('type : ', mat3.dtype)
pixel = mat3[0, 0]
print('pixel : ', pixel)
blue = mat3[0, 0, 0]
green = mat3[0, 0, 1]
red = mat3[0, 0, 2]
print('red : ', red, '   green : ', green, '   blue : ', blue)
for i in range(300):
    mat3.itemset((i,i,0),255)
    mat3.itemset((i,i,1),0)
    mat3.itemset((i,i,2),255)
roi = mat3[0:50,0:100]
mat3[50:100,0:100] = roi
b1 = mat3[:,:,0]
g1 = mat3[:,:,1]
r1 = mat3[:,:,2]
b,g,r = cv2.split(mat3)
bgr3 = cv2.merge((b,g,r))
bgr1 = cv2.cvtColor(bgr3, cv2.COLOR_RGB2GRAY)
bgr13 = cv2.cvtColor(bgr1, cv2.COLOR_GRAY2BGR)
small = cv2.resize(mat3, (30, 30))
mat3 = cv2.line(mat3, (300, 0), (0, 300), (0, 255, 0), 5)
mat3 = cv2.rectangle(mat3, (200, 200), (300, 300), (0, 255, 0), 5)
mat3 = cv2.circle(mat3, (150, 150), 100, (0, 255, 0), 5)
```
