```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
gray = cv2.imread("/work/wangwenkai40/OpenCV/image/lena.png", cv2.IMREAD_GRAYSCALE)
showGray(gray)
```

cv2.THRESH_BINARY

```
ret, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
print(ret)
showGray(thresh1)
```

```
ret, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
print(ret)
showGray(thresh1)
```
