```
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```
color = cv2.imread("/work/f103/coin_8_6x5.jpg", cv2.IMREAD_COLOR)
color.shape
```

```
color = cv2.imread("/work/f103/coin_8_6x5.jpg", cv2.IMREAD_COLOR)
cv2.circle(color, (1900, 650), 550, (0, 255, 0), 20)
cv2.line(color, (1900, 650), (1900, 650 + 550), (0, 255, 0), 10)
cv2.line(color, (1900, 650), (1900, 650 - 550), (0, 255, 0), 10)
cv2.line(color, (1900, 650), (1900 + 550, 650), (0, 255, 0), 10)
cv2.line(color, (1900, 650), (1900 - 550, 650), (0, 255, 0), 10)
cv2.line(color, (1900 - 400, 650 - 200), (1900 + 400, 650 - 200), (0, 255, 0), 10)
cv2.line(color, (1900 - 400, 650 + 200), (1900 + 400, 650 + 200), (0, 255, 0), 10)
cv2.line(color, (1900 - 400, 650 - 200), (1900 - 400, 650 + 200), (0, 255, 0), 10)
cv2.line(color, (1900 + 400, 650 - 200), (1900 + 400, 650 + 200), (0, 255, 0), 10)
showColor(color)
```

```
import numpy as np
rows,cols,ch = color.shape
pts1 = np.float32([[1900+400,650-200],[1900-400,650-200],[1900-400,650+200]])
pts2 = np.float32([[0,0],[0,800],[400,800]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(color,M,(400,800))
showColor(dst)
```

```
M = cv2.getRotationMatrix2D([1900,650], 90, 1)
dst = cv2.warpAffine(color,M,(cols,rows))
dst1 = dst[650-400:650+400,1900-200:1900+200]
showColor(dst1)
```

```
pts1 = np.float32([[1900+400,650-200],[1900-400,650-200],[1900-400,650+200],[1900+400,650+200]])
pts2 = np.float32([[0,0],[0,800],[400,800],[400,0]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(color,M,(400,800))
showColor(dst)
```
