```
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
shui = [172.8, 151.3, 194.3, 215.8, 159.6, 172.4]
dian = [174.92, 198.5, 177.66, 169.99, 153.54, 265.89, 
        286.64, 578.82, 373.66, 520.54, 300.21, 166.7]
qi = [94.95, 127.68, 91.67, 101.49, 94.95, 104.77]
font = FontProperties(fname=r"/work/msyh.ttc", size=14)
labels = '水', '电', '煤气'
sizes = [sum(shui), sum(dian), sum(qi)]
explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
pie = ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.set_label(labels)
ax1.axis('equal')
plt.title("2021年水电气比例图", fontproperties=font)
for front in pie[1]:
        front.set_fontproperties(font)
plt.show()
```
