```python
import cv2
import numpy as np
from PIL import ImageGrab
from matplotlib import pyplot as plt
from pynput.mouse import Button
from pynput.mouse import Controller as shubiao
from pynput.keyboard import Key
from pynput.keyboard import Controller as jianpan

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
```

```python
mouse = shubiao()
```

```python
im = ImageGrab.grab(bbox=(0,0,1366,768))
color = np.array(im)
print(mouse.position)
cv2.line(color, (mouse.position[0], 0), (mouse.position[0], 768), (0, 255, 0), 5)
cv2.line(color, (0, mouse.position[1]), (1366, mouse.position[1]), (0, 255, 0), 5)
showColor(color)
```
